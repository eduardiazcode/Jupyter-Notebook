```python
import pandas as pd
# Leer datos desde archivo CSV
csv_path = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
# Guarga el archivo csv en la variable csv_path
df=pd.read_csv(csv_path,sep=",",header=None)
# df es data frame = es como una tabla de datos lo mas comun es que se use df
# csv_path = indicamos de donde debe leer el archivos
# sep"," = le decimos que la informacion esta separado por comas
# header=None = estamos diciendo que los datos no tienen encabezados
df.head()
# muestra las primeras 5 filas del dataframe
```

```python
# crear la lista headers
headers = ["age","workclass","fnlwgt","education","education-num", "marital-status","occupation",
         "relationship","race","sex", "capital-gain","capital-loss","hours-per-week","native-country","makes"]
df.columns = headers
df.head(100)
```

```python
#La función dtypes genera una tabla con el tipo de dato de cada columna
df.dtypes
```

```python
# Consultado el tipo de dato de un columna especifica (educacion)
print(df['education'].dtype)
print(df['age'].dtype)
print(df['sex'].dtype)
```

```python
# Para saber la cantidad de atributos y elementos que se tienen en el conjunto de datos usaremos (df.shape)
# Esta función visualiza primero el número de elementos y luego el número de atributos.
df.shape
```

```python
# Para visualizar las estadisticas del conjunto de datos (df.describe())
# Moastrara el recuento de la variable, la media, desviacion estandar, el valor minimo, el valor maximo, quartiles
df.describe()
```

```python
# Es posible hacer que la función describe funcione también para las columnas de tipo object.
# podríamos añadir un argumento include="object" o include="all" entre los paréntesis de la función describe
df.describe(include=['object'])
```

```python
# Utiliza la función describe para visualizar la información de todos los atributos:
df.describe(include="all")
```

```python
# Resumir los datos categoricos utilizando la funcion value_counts
education_counts = df['education'].value_counts().to_frame()
education_counts
```
