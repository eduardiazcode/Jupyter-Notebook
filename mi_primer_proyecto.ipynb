```python
import pandas as pd
```

```python
# Leer datos desde archivo CSV
csv_path = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
# Guarga el archivo csv en la variable csv_path
df=pd.read_csv(csv_path,sep=",",header=None)
# df es data frame = es como una tabla de datos lo mas comun es que se use df
# csv_path = indicamos de donde debe leer el archivos
# sep"," = le decimos que la informacion esta separado por comas
# header=None = estamos diciendo que los datos no tienen encabezados
df.head()
# que muestre las primeras 5 filas del dataframe
```

```python
# Nombre de los atributos
# Crear una lista con los encabezados que tendras las columnas header (los encabezados)
headers = ["age", "workclass","fnlwgt","education","education-num", "marital-status","occupation",
         "relationship","race","sex", "capital-gain","capital-loss","hours-per-week","native-country","makes"]
df.columns = headers
# asigna esa lista a las columnas del dataframe
df.head(100)
# que muestre las 100 primeras filas
```

```python
#La función dtypes genera una tabla con el tipo de dato de cada columna
df.dtypes
```
